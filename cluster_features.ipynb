```
import pandas as pd 
import numpy as np

from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```
df = pd.read_csv('features.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```
df.shape
```

```
fdf = df.drop(columns=['emotion'])
fdf.head()
```

```
tdf = fdf.T
tdf
```

**CLUSTERING FEATURES: K-Means Clustering:**  
K-Means Clustering with K=2 and K=3

First, with K = 2:

```
from sklearn.cluster import KMeans
# K=2
k2means = KMeans(n_clusters=2, random_state=42).fit(tdf)
k2means.labels_
```

```
k2means.cluster_centers_
```

Clearly, the clustering is very skewed as 6 out of 7 features have been clustered into one, while only one feature constitutes the second cluster. However, We will still try to classify based on this. 

Saving the data from the features taken from above

```
k2df = pd.DataFrame(data=k2means.cluster_centers_, index=['f1','f2'])
k2df = k2df.T
k2df
```

```
k2df = k2df.join(df['emotion'])     # Join the emotion labels
k2df
```

```
k2df.to_csv('k2means_eye.csv')
```

Similarly, using k = 3:

```
k3means = KMeans(n_clusters=3, random_state=42).fit(tdf)
k3means.labels_
```

```
k3means.cluster_centers_
```

Again it is seen that AU45 is in it's own cluster, while the other features are divided into two clusters in a 1:2 ratio. We will save the data from this clustering as well.

```
k3df = pd.DataFrame(data=k3means.cluster_centers_, index=['f1','f2', 'f3'])
k3df = k3df.T
k3df = k3df.join(df['emotion'])     # Join the emotion labels
k3df
```

```
k3df.to_csv('k3means_eye.csv')
```
